# Dealing with missing values

## Missing values in Age column

```python
sns.boxplot(x='Pclass',y='Age', data=df);

print(df.groupby('Pclass')['Age'].median())
```

```python
sns.heatmap(df.corr(), annot=True)
```

Heatmap shows, passenger class and age are moderately related to each other in inverse way. When we plot boxplot for Age for each passenger class, we can get relation between age and class. It is observed that passengers travelled in class 1 is of higher age group. For class 1,2 and 3 meadian age of passengers are 37, 29 and 24 respectively. 

As we now have the information about relation between age and class, we will replace the missing value of age for a given class with the median age value representing that class.

```python
# Creating copy of the dataset
df_copy= df[['Pclass', 'Age']]
```

```python
for i in (df_copy.index):
    
    if math.isnan(df_copy.loc[i, 'Age']):
        
        # for class 1
        if df_copy.loc[i,'Pclass']==1:
            df_copy.loc[i,'Age']= 37
            
        # for class 2
        elif df_copy.loc[i,'Pclass']==2:
            df_copy.loc[i,'Age']= 29
            
        # for class 3
        else:
            df_copy.loc[i, 'Age']=24
```

```python
# Replacing age column in df with new one
df['Age']= df_copy['Age']
```

## Missing values in Cabin Column

Since 77% of cabin entries are missing, it will not make sense to replace these values with mode. So we will completely drop the cabin column

```python
df.drop(columns=['Cabin'], inplace=True)
```

## Missing values in Embarked Column

Only 2 entries are missing from the Embarked column. We will replace missing values with mode, i.e. port S

```python
mode= df.Embarked.mode()
df['Embarked'].fillna(mode[0], inplace=True)
```

```python
# validating that missing values are removed

df.isna().sum()
```

# Dealing with outliers

```python
sns.catplot(y= 'Fare', data=df, kind='box');
```

Some passengers paid over 500 USD which is much higher than what other passengers paid. It is pulling the mean value to the upper limit. We will drop the row where the Fare is above 500 USD. There are also other outliers but we will keep them as is to avoid adverse effects on other variable after dropping them

```python
# getting index values for rows where fare is more than 500 USD
index= df[df['Fare']>500].index


# dropping rows with index values
df.drop(index= index, inplace =True)
```

# Modifying DataSet

## Adding new columns in the dataset

We have two columns SibSp (Siblings and Spouse) and Parch(Parents and Children) in the dataset. These columns if combined forms the strength of the family of a passenger. let's create a new column Family representing the strength of the passengers' families.

```python
df['Family']= df['SibSp'] + df['Parch']
df.head(3)
```

We will drop column Ticket as it will not be relevant in our further analysis

```python
df.drop(columns=['Ticket'], inplace=True)
df.head(3)
```

Let's also add another column indicating whether a passenger is traveling 'alone' or in 'group'. If the value in 'Family' is 0 then we append 'alone' otherwise, we will append 'group'.

```python
df['Travelling']=df['Family'].apply(lambda x: 'alone' if x==0 else 'group')
df.head()
```
